````python
%%time
# learn equation
equations = pysr(X_np,
                Y_np_flat,
                loss="LPDistLoss{2}()",
                procs=10,
                populations=5,
                niterations=5,
                binary_operators=["plus", "sub", "mult", "pow", "div"],
                unary_operators=[],
                verbosity=0, 
                variable_names=['WOB', 'Db', 'RPM', 'UCS', 'NOC', 'BR', 'SR', 'Dc', 'NOB'])
````

Define a function and edit it with the best function found by pysr library.

````python
def best_equation(row):
    WOB, Db, RPM, UCS, NOC, BR, SR, Dc, NOB = row
#     Y= (pow((((((RPM + 98.457596) / Dc) * WOB) / (UCS - WOB)) - NOB) - NOC, 1.1348255) * 0.31298584)
    Y= (((((((RPM - -51.923843) / Dc) - pow(NOC, 1.5514544)) * WOB) * 1.3547826) / UCS) - (RPM * 0.22286956))
    return Y
````

### Compare the results of PDC ROP model found by pysr with the data.

````python
Data_points_index = np.array(range(0,len(Y_np)))
estimated = np.apply_along_axis(best_equation, 1, X_np)
plt.figure(figsize=(18,7))
plt.scatter(Data_points_index, estimated, label="Best PDC ROP Model found by AI",c='r')
plt.scatter(Data_points_index, Y_np, label="Data")
plt.title("Comparison between PDC ROP Model found by AI and ROP DATA")
plt.xlabel("Data point index")
plt.ylabel("ROP Data (ft/hr)")
plt.legend()
# plt.show()
sys.path.insert(1, '../OpenSource_ROP_Models/ROP models/AI-to-develope-drilling-ROP-model/')
plt.savefig("../AI-to-develope-drilling-ROP-model/Figures/Comparison between PDC ROP Model found by AI and ROP DATA.png", \
            dpi=100)
````

````python
def func(x, a):
    return (a * x)
popt, pcov = curve_fit(func, estimated, Y_np_flat)

plt.figure(figsize=(7,7))
plt.scatter(estimated, Y_np, c='k', label='Data')
plt.xlim(0,120),plt.xlabel("PDC ROP Model estimation (ft/hr)")
plt.ylim(0,120),plt.ylabel("ROP Data (ft/hr)")
plt.plot(estimated, func(estimated, *popt), 'r-',
         label='fit: slope = %5.6f' % tuple(popt))
plt.legend()
plt.savefig("../AI-to-develope-drilling-ROP-model/Figures/ModelvsData.png")
````

````python
print("Mean Squared Error: ", np.mean((Y_np_flat-func(estimated, *popt))**2))
````

### Sensitivity Analysis on Model

````python
problem = {
    'num_vars': 5,
    'names': ['RPM', 'Dc', 'NOC', 'WOB', 'UCS'],
    'bounds': [[10, 300],
               [0.3, 0.7],
               [8, 16],
              [100,5000],
              [5000,25000]]
}
````

````python
def evaluate_model(row):
    RPM, Dc, NOC, WOB, UCS = row
    Y= (((((((RPM - -51.923843) / Dc) - pow(NOC, 1.5514544)) * \
           WOB) * 1.3547826) / UCS) - (RPM * 0.22286956))
    return Y
````

````python
problem
````

````python
param_values = saltelli.sample(problem, 1024)

Y = np.zeros([param_values.shape[0]])
for i, X in enumerate(param_values):
    Y[i] = evaluate_model(X)
````

````python
Si = sobol.analyze(problem, Y, print_to_console=False)
ax = Si.plot()
# plt.savefig("../AI-to-develope-drilling-ROP-model/Figures/Sensitivity_analysis_1.png")
````

````python
Si_filter = {k:Si[k] for k in ['ST','ST_conf','S1','S1_conf']}
Si_df = pd.DataFrame(Si_filter, index=problem['names'])
Si_df
````

````python
fig, ax = plt.subplots(1)
indices = Si_df[['ST','ST_conf','S1','S1_conf']]
err = Si_df[['ST','ST_conf','S1','S1_conf']]

indices.plot.bar(yerr=err.values.T,ax=ax)
fig.set_size_inches(8,4)

# plt.show()
# plt.savefig("../AI-to-develope-drilling-ROP-model/Figures/Sensitivity_analysis2.png")
````

Based on sensitivity analysis one can see that the main parameters are bit diameter, weight on bit, RPM, rock strength, number of cutters. 

It is also seen the second-order interaction between independent parameters. It is seen that the WOB, UCS and RPM have more contribution to the ROP compare to NOC and Dc.It is also seen that mathematically there is a big interaction between RS and bit diameter, WOB-UCS and WOB-RPM.

Note that these interactions are from mathematical perspective and does not necessarily represent physics.
